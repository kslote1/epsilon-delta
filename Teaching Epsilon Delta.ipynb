# Epsilon-Delta Proof for f(x)

In this notebook, we'll demonstrate the epsilon-delta proof for the function $f(x)$. We'll find the limit as $x$ approaches a certain value, say $a = 3$.

## Introduction to Epsilon-Delta Definition

The epsilon-delta definition of the limit for a function $f(x)$ is given by:

$$\lim_{x \to a} f(x) = L$$

if for every positive number $\epsilon > 0$, there exists a positive number $\delta > 0$ such that if $0 < |x - a| < \delta$, then $|f(x) - L| < \epsilon$.

## Choosing the Function f(x) = 2x + 1

For our example, we'll use the function:

$$f(x) = 2x + 1$$

We'll find the limit as $x$ approaches $3$, i.e., $a = 3$.

## Implementing the Epsilon-Delta Proof in Python

We'll define a function `epsilon_delta_proof` that takes $a$, $L$, and $\epsilon$ as input arguments and checks if the epsilon-delta condition is met for the given function:

```python
def epsilon_delta_proof(a, L, epsilon):
    delta = epsilon / 2
    x_values = [a - delta, a + delta]
    y_values = [f(x) for x in x_values]
    
    return all(abs(y - L) < epsilon for y in y_values)
```

## Visualizing the Proof using Matplotlib

We'll use Matplotlib to visualize the function $f(x) = 2x + 1$, the limit $L$ as $x$ approaches $3$, and the epsilon and delta ranges to demonstrate the epsilon-delta proof visually.
```

````python
import numpy as np

def f(x):
    return 2 * x + 1

def epsilon_delta_proof(a, L, epsilon):
    delta = epsilon / 2
    x_values = [a - delta, a + delta]
    y_values = [f(x) for x in x_values]
    
    return all(abs(y - L) < epsilon for y in y_values)

a = 3
L = f(a)
epsilon = 0.1

proof_result = epsilon_delta_proof(a, L, epsilon)
print("Epsilon-delta proof result:", proof_result)
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(a - 2, a + 2, 1000)
y = f(x)

plt.plot(x, y, label='f(x) = 2x + 1')

delta = epsilon / 2
x_interval = [a - delta, a + delta]
y_interval = [L - epsilon, L + epsilon]

plt.fill_between(x_interval, y_interval[0], y_interval[1], color='gray', alpha=0.3, label='Epsilon range')
plt.axvspan(x_interval[0], x_interval[1], color='blue', alpha=0.3, label='Delta range')

plt.axhline(y=L, color='red', linestyle='--', label=f'Limit L = {L}')
plt.axvline(x=a, color='green', linestyle='--', label=f'x = {a}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Epsilon-Delta Proof Visualization')
plt.grid(True)
plt.show()
````
